```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_file = "./Resources/athletic_sales_2020.csv"
sales_2021_file = "./Resources/athletic_sales_2021.csv"

sales_2020 = pd.read_csv(sales_2020_file)
sales_2021 = pd.read_csv(sales_2021_file)
```

```python
# Display the 2020 sales DataFrame
sales_2020
```

```python
# Display the 2021 sales DataFrame
sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
Joined_data =  pd.concat([sales_2020, sales_2021] ,axis = "rows", join = "inner")
Joined_data = Joined_data.reset_index(drop= True)
Joined_data.head(30)
```

```python
# Check if any values are null.
Joined_data.isnull().sum()
```

```python
# Check the data type of each column
Joined_data.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
Joined_data["invoice_date"] = pd.to_datetime(Joined_data["invoice_date"], errors = "coerce")
```

```python
# Confirm that the "invoice_date" data type has been changed.
Joined_data.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouping_product = Joined_data.groupby(["region", "state", "city"])[["units_sold"]].agg(["sum"])

grouped_products = grouping_product.rename(columns={"sum": "Total_Products_Sold"})

# Show the top 5 results.
grouped_products     
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold = pd.pivot_table(Joined_data, index=["region", "state", "city"], values = 'units_sold', aggfunc= "sum")

# Rename the "units_sold" column to "Total_Products_Sold"
Products_sold_renamed = products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
Products_sold_renamed.sort_values(by= "Total_Products_Sold", ascending= False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear = Joined_data.drop(Joined_data[Joined_data["product"] != "Women's Athletic Footwear"].index)
#womens_footwear = Joined_data[Joined_data["product"] == "women's atheltic footwear"]
womens_footwear.dtypes
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_grouped = womens_footwear.groupby(["retailer", "region", "state", "city"])[["units_sold"]].agg(["sum"])
# Show the top 5 results.
#womens_footwear_grouped.sort_values(by = "units_sold", ascending= False)
womens_footwear_grouped.units_sold.sort_values(by = "sum", ascending= False)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_pivot = pd.pivot_table(womens_footwear, index = ["retailer", "region", "state", "city"], values= ["units_sold"], aggfunc= "sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_renamed = womens_footwear_pivot.rename(columns = {"units_sold": "Womens_Footwear_Units_sold"})
# Show the top 5 results.
womens_footwear_renamed.sort_values(by = "Womens_Footwear_Units_sold",ascending= False)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
Pivot_date_sales = pd.pivot_table(womens_footwear, index = ["invoice_date"], values = ["total_sales"], aggfunc= "sum")

# Optional: Rename the "total_sales" column to "Total Sales"
Pivot_date_sales_renamed =  Pivot_date_sales.rename(columns = {"total_sales": "Total Sales"})

# Show the table.
Pivot_date_sales_renamed
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_resampled_daily = Pivot_date_sales_renamed.resample("D").sum().round(1)

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_resampled_daily_ordered = pivot_resampled_daily.sort_values(by = "Total Sales", ascending = False)
pivot_resampled_daily_ordered
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_resampled_weekly = Pivot_date_sales_renamed.resample("W").sum().round(1)

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_resampled_weekly.sort_values(by = "Total Sales", ascending = False)
```
